```python
#꽃 종류 구분 through classification
from sklearn.datasets import load_iris
import pandas as pd

#load_iris :아이리스 꽃 데이터 set
```

```python
#since this data have the label which is target, it is supervised machine learning
iris = load_iris()
#feature data
iris.data
#label (target) data
iris.target
```

```python
#based on sepal length and width and petal length and width, expect the 꽃의 품종 which are setosa, vesicolor, and virginica

#steps to do the machine learning of it
# 1. classify the data set (학습 데이터 and 테스트 데이터 분리)
from sklearn.model_selection import train_test_split

# x= feature and y = label(target) // train = 학습 and test = 테스트
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=123)

y_test
len(y_train)
```

```python
# 2. model learning (based on the 학습 데이터, 모델 학습시키기 by applying the machine learning algorithm)
from sklearn.tree import DecisionTreeClassifier

dt_clf = DecisionTreeClassifier(random_state=123)
dt_clf
```

```python
# 3. 예측 수행 (predict the test data)
dt_clf.fit(x_train, y_train)
```

```python
# 4.평가 (by comparing the 결과값 and 데이터 결과값, 정확도 평가)
predict = dt_clf.predict(x_test)
predict
```

```python
#5. 채점 진행
from sklearn.metrics import accuracy_score

accuracy_score(y_test, predict)
```

```python
df = pd.DataFrame(data=x_test, columns=iris.feature_names)
df["answer"]=y_test
df["predict"]=predict

df
```
