[문제2]

Python의 얕은 복사와 깊은 복사에 대해서 설명하세요

A shallow copy constructs a new compound object and then (to the extent possible) inserts references into it to the objects found in the original.
It doesn't effect the original and copied one since it make new id for copy. But it can be issued when you have list within list that both of lists are mutable. Because you will need to change the values for both lists when you need to mutate the value of original list.

A deep copy constructs a new compound object and then, recursively, inserts copies into it of the objects found in the original. It copies lists within list. So, you don't need to changethe mutable value when you need to change the values in original list. But this means the revise of original will effect the copied one.
